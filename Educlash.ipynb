### Récupération des données des lycées

Dans cette cellule, nous récupérons via une API les données intitulées **"Indicateurs de valeur ajoutée des lycées d'enseignement général et technologique (ancien)"**. Ces données nous serviront à analyser les performances des lycées publics et privés sous contrat.


```python
url_lycee_ancien = "https://data.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-indicateurs-de-resultat-des-lycees-denseignement-general-et-technologique/records?offset=9000&limit=10"
res = requests.get(url_lycee_ancien)
if res.status_code == 200 :
  data = res.json()
  print("Données récupérées avec succès")
  print(data)
  print("taille : "+str(len(data['results'])))
else :
  print("Erreur :  l'accès à l'API a échoué")
```

```python
df = pd.DataFrame(data['results'])
df
```

```python
limit = 10  # Nombre de résultats par page

# Boucle sur les deux premières pages
for page in range(1, 3):  # On veut 20 résultats, donc 2 pages
    params = {"page": page, "limit": limit}
    response = requests.get(url_lycee_ancien, params=params)
    print(page)
    print(response.url)
```
