### Récupération des données des lycées

Dans cette cellule, nous récupérons via une API les données intitulées **"Indicateurs de valeur ajoutée des lycées d'enseignement général et technologique (ancien)"**. Ces données nous serviront à analyser les performances des lycées publics et privés sous contrat.


```python
url_lycee_ancien = "https://data.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-indicateurs-de-resultat-des-lycees-denseignement-general-et-technologique/records?limit=20"
response = requests.get(url_lycee_ancien)

if response.status_code == 200:
    dico_lycee_ancien = response.json()
    print(dico_lycee_ancien)
else:
    print("Erreur lors de la récupération des données")
```

Conversion du dictionnaire lycee ancien vers dataframe panda

```python
df_lycee_ancien = pd.json_normalize(dico_lycee_ancien['results'])
df_lycee_ancien.head()
```
